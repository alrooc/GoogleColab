```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
```

```python
def plotVectors(vecs, cols, alpha=1):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)
    ax.set_aspect('equal', adjustable='box')

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
```

```python
def graficar_matriz(matrix):

    origin2D = np.vstack([0,0])
    fig = plt.figure()
    ax = fig.add_subplot(111)


    # c√≠rculo unitario
    x = np.linspace(-1, 1, 100000)
    y = np.sqrt(1-(x**2))


    # circulo modificado (separando las partes positivas y negativas)
    x1 = matrix[0,0]*x + matrix[0,1]*y
    y1 = matrix[1,0]*x + matrix[1,1]*y
    x1_neg = matrix[0,0]*x - matrix[0,1]*y
    y1_neg = matrix[1,0]*x - matrix[1,1]*y

    # Vectores
    u1 = [matrix[0,0],matrix[1,0]]
    v1 = [matrix[0,1],matrix[1,1]]
    plt.quiver(*origin2D, *u1, angles='xy', scale_units='xy', scale=1,color="gold")
    plt.quiver(*origin2D, *v1, angles='xy', scale_units='xy', scale=1,color="red")
    plt.plot(x1, y1, 'g', alpha=0.5)
    plt.plot(x1_neg, y1_neg, 'g', alpha=0.5)


    ax.set_aspect('equal', adjustable='box')
```

```python
A = np.array([[4, 7], [6, 1]])


print ('circulo unitario:')
graficar_matriz(np.array([[1, 0], [0, 1]]))
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

print ('Unit circle transformed by A:')


graficar_matriz(A)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

## Ejemplo del Tema 4

Descomponer la matriz $A$

```python
A = np.array([[3, 7], [5, 2]])
print(A)
```

Paso 1: Encontrar las 3 matrices que la componen:

```python
U, D, V = np.linalg.svd(A)

print(U)
print(D)
print(V)
```

```python
graficar_matriz(V)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
np.diag(D)
```

```python
graficar_matriz(np.diag(D).dot(V))
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

```python
graficar_matriz(U.dot(np.diag(D)).dot(V))
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
u1 = [D[0]*U[0,0], D[0]*U[0,1]]
v1 = [D[1]*U[1,0], D[1]*U[1,1]]
origin2D = np.vstack([0,0])



graficar_matriz(U.dot(np.diag(D)).dot(V))

plt.text(-5, -4, r"$\sigma_1u_1$", size=18)
plt.text(-4, 1, r"$\sigma_2u_2$", size=18)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.quiver(*origin2D, *u1, angles='xy', scale_units='xy', scale=1,color="k")
plt.quiver(*origin2D, *v1, angles='xy', scale_units='xy', scale=1,color="k")
plt.show()
```
